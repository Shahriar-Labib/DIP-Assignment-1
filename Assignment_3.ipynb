```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img1 = np.zeros((400, 400, 3), dtype=np.uint8)
img2 = np.zeros((400, 400, 3), dtype=np.uint8)

cv2.rectangle(img1, (100, 150), (300, 250), (0, 255, 0), -1)

cv2.circle(img2, (200, 200), 100, (255, 0, 255), -1)

def display_images(images, titles, ncols=3):
    n_images = len(images)
    nrows = (n_images + ncols - 1) // ncols
    plt.figure(figsize=(15, 5 * nrows))

    for i, img in enumerate(images):
        plt.subplot(nrows, ncols, i + 1)
        plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
        plt.title(titles[i], fontsize=12)
        plt.axis('off')

    plt.tight_layout()
    plt.show()
```

```python
# Bitwise AND Task 1
bitwise_and = cv2.bitwise_and(img1, img2)

# Bitwise OR Task 2
bitwise_or = cv2.bitwise_or(img1, img2)

# Bitwise XOR Task 3
bitwise_xor = cv2.bitwise_xor(img1, img2)

# Bitwise NOT Task 4
bitwise_not1 = cv2.bitwise_not(img1)
bitwise_not2 = cv2.bitwise_not(img2)

# Element-wise Multiplication Task 5
elementwise_multiplication = cv2.multiply(img1, img2)

# Alpha Blending (with some transparency) Task 6
alpha_blending = cv2.addWeighted(img1, 0.7, img2, 0.3, 0)

# Image Subtraction Task 7
image_subtraction = cv2.subtract(img1, img2)
```

```python
display_images([bitwise_and, bitwise_or, bitwise_xor, bitwise_not1, bitwise_not2,
                elementwise_multiplication, alpha_blending, image_subtraction],
               ["AND", "OR", "XOR", "NOT img1", "NOT img2", "Multiplication", "Alpha Blending", "Subtraction"])
```
