```python
#2125051028
import cv2
import matplotlib.pyplot as plt
import numpy as np
import urllib.request
from PIL import Image
```

```python
url = 'https://plus.unsplash.com/premium_photo-1680100255906-09ffdc2eeaaf?q=80&w=2070&auto=format&fit=crop&ixlib=rb-4.0.3&ixid=M3wxMjA3fDB8MHxwaG90by1wYWdlfHx8fGVufDB8fHx8fA%3D%3D'  # Replace with your image URL

resp = urllib.request.urlopen(url)
img_np = np.asarray(bytearray(resp.read()), dtype=np.uint8)
img = cv2.imdecode(img_np, cv2.IMREAD_COLOR)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
# Resize to 256x256
resized = cv2.resize(img, (256, 256))

# Rotate 180 degrees clockwise
rotated = cv2.rotate(resized, cv2.ROTATE_180)

# Vertical flip
flipped = cv2.flip(resized, 0)

# Plot all images
titles = ['Original', 'Resized (256x256)', 'Rotated 180Â°', 'Vertical Flip']
images = [img, resized, rotated, flipped]
```

```python
plt.figure(figsize=(12, 6))
for i in range(4):
    plt.subplot(1, 4, i + 1)
    plt.imshow(images[i])
    plt.title(titles[i])
    plt.axis('off')

plt.tight_layout()
plt.show()
```
