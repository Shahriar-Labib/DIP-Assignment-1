```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from urllib.request import urlopen
import numpy as np
```

```python
url = 'https://images.unsplash.com/photo-1743083931860-05b751559acb?q=80&w=2070&auto=format&fit=crop&ixlib=rb-4.0.3&ixid=M3wxMjA3fDB8MHxwaG90by1wYWdlfHx8fGVufDB8fHx8fA%3D%3D'  # Replace with any image URL
resp = urlopen(url)
image_array = np.asarray(bytearray(resp.read()), dtype=np.uint8)
image = cv2.imdecode(image_array, cv2.IMREAD_COLOR)


image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)


plt.imshow(image_rgb)
plt.title("Original Image")
plt.axis("off")
plt.show()
```

```python
height, width, _ = image.shape
center_x, center_y = width // 2, height // 2

half_crop = 50
crop_img = image[
    center_y - half_crop : center_y + half_crop,
    center_x - half_crop : center_x + half_crop
]

crop_rgb = cv2.cvtColor(crop_img, cv2.COLOR_BGR2RGB)

plt.imshow(crop_rgb)
plt.title("Cropped 100x100 Region (Center)")
plt.axis("off")
plt.show()
```

```python
hsv_img = cv2.cvtColor(crop_img, cv2.COLOR_BGR2HSV)
h, s, v = cv2.split(hsv_img)

fig, axs = plt.subplots(1, 3, figsize=(15, 5))
axs[0].imshow(h, cmap='gray')
axs[0].set_title('Hue Channel')
axs[0].axis('off')

axs[1].imshow(s, cmap='gray')
axs[1].set_title('Saturation Channel')
axs[1].axis('off')

axs[2].imshow(v, cmap='gray')
axs[2].set_title('Value Channel')
axs[2].axis('off')

plt.tight_layout()
plt.show()
```
